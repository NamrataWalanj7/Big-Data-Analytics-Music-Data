```
from pyspark.sql import SparkSession
```

```
spark = SparkSession.builder.appName('Project').getOrCreate()
```

```
df = spark.read.csv('/FileStore/tables/finaldata.csv', header = True, inferSchema = True)
```

```
df.printSchema()
```

```
#df = df.na.drop()
```

```
grouped_data = df.groupBy(['genre'])
grouped_data1 = df.groupBy(['genre'])
grouped_data2 = df.groupBy(['genre'])
```

```
grouped_data.agg({'genre': 'count'}). show()
```

```
grouped_data1.agg({'track_listens': 'avg'}). show()
```

```
grouped_data2.agg({'duration': 'avg'}). show()
```

```
df_genre = grouped_data.agg({'genre': 'count'})
df_genre = df_genre.withColumnRenamed('count(genre)','genre_count')
df_genre.show()
```

```
df_genre_listens = grouped_data1.agg({'track_listens': 'avg'})
df_genre_listens = df_genre_listens.withColumnRenamed('avg(track_listens)','genre_listens')
df_genre_listens.show()
```

```
df_genre_duration = grouped_data2.agg({'duration': 'avg'})
df_genre_duration = df_genre_duration.withColumnRenamed('avg(duration)','genre_duration')
df_genre_duration.show()
```

```
from pyspark.sql.types import IntegerType
df_genre = df_genre.withColumn('genre_count', df_genre['genre_count'].cast(IntegerType()))
df_genre.printSchema()

df_genre_listens = df_genre_listens.withColumn('genre_listens', df_genre_listens['genre_listens'].cast(IntegerType()))
df_genre_listens.printSchema()

df_genre_duration = df_genre_duration.withColumn('genre_duration', df_genre_duration['genre_duration'].cast(IntegerType()))
df_genre_duration.printSchema()
```

```
total_rows = df.count()
total_rows
```

```
df_genre = df_genre.withColumn('percent', (df_genre['genre_count']*100)/total_rows)
df_genre.show()
```

```
df_genre_filter = df_genre.filter(df_genre['percent'] >= 4).select(['genre','percent'])
df_genre_filter.show()
```

```
df_genre_listens.createOrReplaceTempView('listens_view')
genre_listens = spark.sql("SELECT * FROM listens_view order by genre_listens desc limit 7")
genre_listens.show()
```

```
df_genre_listens.createOrReplaceTempView('listens_view')
genre_listens_asc = spark.sql("SELECT * FROM listens_view where genre_listens is not null order by genre_listens asc limit 7")
genre_listens_asc.show()
```

```
df_genre_duration.createOrReplaceTempView('duration_view')
genre_duration = spark.sql("SELECT * FROM duration_view where genre_duration is not null order by genre_duration asc")
genre_duration.show()
```

```
pandas_df_genre = df_genre_filter.toPandas()
pandas_df_genre
```

```
pandas_df_listens = genre_listens.toPandas()
pandas_df_listens
```

```
pandas_df_listens_asc = genre_listens_asc.toPandas()
pandas_df_listens_asc
```

```
pandas_df_duration = genre_duration.toPandas()
pandas_df_duration
```
